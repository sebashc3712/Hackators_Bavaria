```python
def AddProductColumn(matrix,df,product):
    
    product=str(product)
    
    matrix['Product']=product
    matrix['Llave']=matrix.period.astype(str)+'-'+matrix.customer.astype(str)+'-'+matrix.Product.astype(str)
    df['Llave']=df.period.astype(str)+'-'+df.Cliente.astype(str)+'-'+df.product_id.astype(int).astype(str)
    
    matrix=matrix.merge(df[['Llave','Volumen','disc','nr','nr/volume']],how='left',on='Llave')
    matrix[['Volumen','disc','nr','nr/volume']]=matrix[['Volumen','disc','nr','nr/volume']].fillna(0)
    matrix=matrix.rename(columns={'Volumen':'Volumen_'+product,'disc':'disc_'+product,'nr':'nr_'+product,
                                 'nr/volume':'nr/volume_'+product})
    matrix=matrix.drop('Product',axis=1)
    
    return matrix
```

```python
def BigMatrix(matrix,df,products):
    
    max_product=max(products.index)
    
    for i in range(0,max_product):
        
        matrix=AddProductColumn(matrix,df,i)
    
    return matrix
```

```python
def lag_feature(df, product, lags, col):
    tmp = df[['period','Client','product_id',col]]
    for i in lags:
        shifted = tmp.copy()
        shifted.columns = ['date_block_num','shop_id','item_id', col+'_lag_'+str(i)]
        shifted['date_block_num'] += i
        df = pd.merge(df, shifted, on=['date_block_num','shop_id','item_id'], how='left')
    return df
```
